#Problem Statements for EDA:

Problem 1: Age and Gender Influence on Heart Disease

Objective: Explore the distribution of heart disease across different age groups and genders. Specifically, investigate whether certain age ranges and one gender are more prone to heart disease than others.

EDA Focus: Analyze the distribution of HeartDisease across Age and Sex. 

Problem 2: Correlation Between Cholesterol, Blood Pressure, and Heart Disease

Objective: Analyze the relationship between cholesterol levels (Cholesterol), resting blood pressure (RestingBP), and the presence of heart disease. Investigate if patients with high cholesterol and/or blood pressure are more likely to suffer from heart disease.

EDA Focus: Utilize scatter plots, correlation matrices, and regression plots to investigate relationships between cholesterol, blood pressure, and heart disease outcomes.


Problem 3: Chest Pain Types and their Association with Heart Disease

Objective: Examine the different types of chest pain (ChestPainType) and their association with heart disease. Explore whether certain types of chest pain are stronger indicators of heart disease compared to others.
EDA Focus: Perform a categorical analysis to examine the frequency of different ChestPainType categories among patients with and without heart disease. Visualize the data using bar plots and pie charts.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv(r"C:\Users\Anurag Kumar\Downloads\archive\heart.csv")

# Problem 1: Age and Gender Influence on Heart Disease
# ------------------------------------------------------
# Creating histograms to analyze the age distribution based on heart disease and gender
plt.figure(figsize=(10, 6))
sns.histplot(df, x='Age', hue='HeartDisease', multiple='stack', kde=False, palette='Set2')
plt.title('Age Distribution by Heart Disease')
plt.show()

# Violin plot to visualize distribution of heart disease by age and gender
plt.figure(figsize=(10, 6))
sns.violinplot(x='Sex', y='Age', hue='HeartDisease', data=df, split=True, palette='Set3')
plt.title('Heart Disease by Age and Gender')
plt.show()

# Problem 2: Correlation Between Cholesterol, Blood Pressure, and Heart Disease
# ---------------------------------------------------------------------------
# Scatterplot to analyze relationship between Cholesterol and RestingBP colored by HeartDisease
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Cholesterol', y='RestingBP', hue='HeartDisease', data=df, palette='coolwarm')
plt.title('Cholesterol vs Resting Blood Pressure Colored by Heart Disease')
plt.show()

# Correlation heatmap focusing on specific features
plt.figure(figsize=(10, 6))
corr_subset = df[['Cholesterol', 'RestingBP', 'HeartDisease']].corr()
sns.heatmap(corr_subset, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Cholesterol, RestingBP, and Heart Disease')
plt.show()

# Problem 3: Chest Pain Types and Association with Heart Disease
# -------------------------------------------------------------
# Bar plot to analyze the relationship between Chest Pain Type and Heart Disease
plt.figure(figsize=(10, 6))
sns.countplot(x='ChestPainType', hue='HeartDisease', data=df, palette='Set1')
plt.title('Chest Pain Types by Heart Disease')
plt.show()

# Pie chart for better visualization of the percentage of chest pain types in heart disease
df.groupby('ChestPainType')['HeartDisease'].value_counts().unstack().plot(kind='pie', subplots=True, autopct='%.1f%%', figsize=(14, 7))
plt.title('Chest Pain Types in Heart Disease')
plt.show()
```

Problem 1: Age and Gender Influence on Heart Disease

We added histograms and violin plots to explore the distribution of age and gender across heart disease patients. The histogram is stacked to show the proportion of heart disease occurrences in different age groups, and the violin plot reveals differences between males and females.
Problem 2: Correlation Between Cholesterol, Blood Pressure, and Heart Disease

A scatterplot was introduced to explore the relationship between cholesterol levels and resting blood pressure, colored by the presence of heart disease. Additionally, a correlation matrix heatmap focusing on Cholesterol, RestingBP, and HeartDisease was generated to examine their relationships numerically.
Problem 3: Chest Pain Types and Association with Heart Disease

A bar plot was used to visualize the distribution of chest pain types and their relationship to heart disease. Additionally, pie charts were introduced to show the proportion of patients with different chest pain types among those with and without heart disease.
Final Report After EDA


Final EDA Report: Heart Disease Dataset
1. Age and Gender Influence on Heart Disease

Findings:
Most patients diagnosed with heart disease are aged between 50 and 60 years.
Gender analysis shows that a higher proportion of males have heart disease compared to females across all age groups.
The violin plot further supports that males have a higher prevalence of heart disease, particularly in middle-aged and older age groups.
2. Correlation Between Cholesterol, Blood Pressure, and Heart Disease

Findings:
The scatterplot did not reveal a strong linear relationship between cholesterol and resting blood pressure. However, it does show that patients with both high cholesterol and high blood pressure are more likely to have heart disease.
The correlation heatmap showed that the correlation between cholesterol, resting blood pressure, and heart disease is weak but positive, suggesting that higher values of these variables are associated with a higher risk of heart disease.
3. Chest Pain Types and Association with Heart Disease

Findings:
The bar plot analysis revealed that patients with typical angina are less likely to be diagnosed with heart disease, while those with asymptomatic chest pain have a significantly higher likelihood of heart disease.
Pie charts showed a clear distinction in the distribution of chest pain types between heart disease patients and non-heart disease patients, with a larger proportion of asymptomatic and atypical angina patients being diagnosed with heart disease.
Conclusions:
This EDA provided insights into the relationships between age, gender, cholesterol, blood pressure, chest pain types, and heart disease. The visualizations revealed that middle-aged males with higher cholesterol and blood pressure, especially those presenting with asymptomatic chest pain, are at greater risk for heart disease. The findings suggest that targeted preventive measures should focus on these groups, especially for routine screenings and early interventions.
